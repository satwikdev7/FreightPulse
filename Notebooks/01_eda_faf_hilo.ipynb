```python
## 1. FAF5.7.1_HiLoForecasts Dataset
```

```python
import os
import pandas as pd

# -------------------------------
# 1. Set file paths
# -------------------------------
base_path = "/Users/satwik/Documents/FreightPulse/Data/FAF5.7.1_HiLoForecasts"

csv_path = os.path.join(base_path, "FAF5.7.1_HiLoForecasts.csv")
metadata_path = os.path.join(base_path, "FAF5_metadata.xlsx")

# -------------------------------
# 2. Load main CSV
# -------------------------------
df = pd.read_csv(csv_path, low_memory=False)

print("=" * 80)
print("DATASET: FAF5.7.1_HiLoForecasts")
print("=" * 80)

# -------------------------------
# 3. Basic structure
# -------------------------------
print("\n1. SHAPE")
print(df.shape)

print("\n2. COLUMN NAMES")
print(df.columns.tolist())

print("\n3. FIRST 5 ROWS")
print(df.head())

print("\n4. LAST 5 ROWS")
print(df.tail())

# -------------------------------
# 4. Data types and info
# -------------------------------
print("\n5. DATA TYPES")
print(df.dtypes)

print("\n6. INFO")
df.info()

# -------------------------------
# 5. Missing values
# -------------------------------
missing_counts = df.isna().sum()
missing_percent = (df.isna().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_percent": missing_percent
}).sort_values(by="missing_count", ascending=False)

print("\n7. MISSING VALUES SUMMARY")
print(missing_summary[missing_summary["missing_count"] > 0])

# -------------------------------
# 6. Duplicates
# -------------------------------
print("\n8. DUPLICATE ROWS")
print(df.duplicated().sum())

# -------------------------------
# 7. Numeric summary
# -------------------------------
print("\n9. NUMERIC SUMMARY")
print(df.describe(include=["number"]).T)

# -------------------------------
# 8. Categorical summary
# -------------------------------
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()

print("\n10. CATEGORICAL COLUMNS")
print(cat_cols)

if cat_cols:
    print("\n11. CATEGORICAL SUMMARY")
    print(df[cat_cols].describe().T)

# -------------------------------
# 9. Unique values per column
# -------------------------------
unique_summary = pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.astype(str),
    "n_unique": [df[col].nunique(dropna=True) for col in df.columns]
}).sort_values(by="n_unique", ascending=False)

print("\n12. UNIQUE VALUE SUMMARY")
print(unique_summary)

# -------------------------------
# 10. Memory usage
# -------------------------------
memory_mb = df.memory_usage(deep=True).sum() / 1024**2
print(f"\n13. MEMORY USAGE: {memory_mb:.2f} MB")

# -------------------------------
# 11. Try loading metadata Excel
# -------------------------------
if os.path.exists(metadata_path):
    print("\n14. METADATA EXCEL FOUND")
    xls = pd.ExcelFile(metadata_path)
    print("Sheet names:", xls.sheet_names)

    for sheet in xls.sheet_names:
        meta_df = pd.read_excel(metadata_path, sheet_name=sheet)
        print(f"\n--- Metadata Sheet: {sheet} ---")
        print("Shape:", meta_df.shape)
        print(meta_df.head())
else:
    print("\n14. METADATA EXCEL NOT FOUND")

# -------------------------------
# 12. Optional: Save summaries
# -------------------------------
missing_summary.to_csv(os.path.join(base_path, "HiLo_missing_summary.csv"))
unique_summary.to_csv(os.path.join(base_path, "HiLo_unique_summary.csv"), index=False)

print("\nDone.")
```
